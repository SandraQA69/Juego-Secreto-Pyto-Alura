### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Cargar cada tienda y agregar una columna de identificación
tiendas = []
for i, url in enumerate(urls, 1):
    df = pd.read_csv(url)
    df['Tienda'] = f'Tienda {i}'
    tiendas.append(df)

# Combinar todos los datos en un solo DataFrame
datos_completos = pd.concat(tiendas, ignore_index=True)

# Mostrar las primeras filas del conjunto combinado
datos_completos.head()
```

```python
# Información básica sobre los datos
print("Información del dataset:")
print(datos_completos.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(datos_completos.describe())

# Verificar valores nulos
print("\nValores nulos por columna:")
print(datos_completos.isnull().sum())
```

#1. Análisis de facturación



```python
# Crear la columna "Ingresos" (Precio + Costo de Envío)
datos_completos["Ingresos"] = datos_completos["Precio"] + datos_completos["Costo de envío"]

# Agrupar por tienda y calcular la facturación total
facturacion_por_tienda = datos_completos.groupby("Tienda")["Ingresos"].sum().reset_index()

# Mostrar la facturación total de cada tienda
print(facturacion_por_tienda)
```

```python
import matplotlib.pyplot as plt

# Gráfico de facturación total por tienda
facturacion_por_tienda = datos_completos.groupby("Tienda")["Ingresos"].sum().reset_index()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(facturacion_por_tienda["Tienda"], facturacion_por_tienda["Ingresos"], color='skyblue')
plt.title("Facturación Total por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Facturación Total")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Ingresos por categoría
ingresos_por_categoria = datos_completos.groupby("Categoría del Producto")["Ingresos"].sum().reset_index()

# Ordenar las categorías por ingresos
ingresos_por_categoria = ingresos_por_categoria.sort_values(by="Ingresos", ascending=False)

# Mostrar las categorías más populares
print(ingresos_por_categoria)
```

```python
# Gráfico de categorías más populares
ingresos_por_categoria = datos_completos.groupby("Categoría del Producto")["Ingresos"].sum().reset_index()
ingresos_por_categoria = ingresos_por_categoria.sort_values(by="Ingresos", ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(ingresos_por_categoria["Categoría del Producto"], ingresos_por_categoria["Ingresos"], color='lightcoral')
plt.title("Categorías Más Populares")
plt.xlabel("Categoría del Producto")
plt.ylabel("Ingresos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de categorías más populares
ingresos_por_categoria = datos_completos.groupby("Categoría del Producto")["Ingresos"].sum().reset_index()
ingresos_por_categoria = ingresos_por_categoria.sort_values(by="Ingresos", ascending=False)

# Crear el gráfico circular
plt.figure(figsize=(8, 8))
plt.pie(ingresos_por_categoria["Ingresos"], labels=ingresos_por_categoria["Categoría del Producto"], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title("Distribución de Ingresos por Categoría")
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.show()
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos
